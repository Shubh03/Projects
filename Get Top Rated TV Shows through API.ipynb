## <center>Extract Data Using API

```python
import pandas as pd
import requests
```

```python
response = requests.get('https://api.themoviedb.org/3/tv/top_rated?api_key=9d2be5724ef73e7e03e0cd32e77bf280&language=en-US&page=1')
```

```python
response
```

```python
temp_df = pd.DataFrame(response.json()['results'])[['id','name','overview','popularity','vote_average','vote_count']]
```

```python
temp_df.head()
```

```python
df=pd.DataFrame()
```

```python
for i in range(1,126):
    response = requests.get('https://api.themoviedb.org/3/tv/top_rated?api_key=9d2be5724ef73e7e03e0cd32e77bf280&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id','name','overview','original_language','first_air_date','popularity','vote_average','vote_count']]
    df = pd.concat([temp_df,df],ignore_index=True)
```

```python
df.rename(
    columns={"name":"Title",
                "first_air_date":"Release_Date"}
          ,inplace=True)
```
